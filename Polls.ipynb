```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 15)
```

```python
# Current cycle polls data
df_polls = pd.read_csv("data/president_polls.csv")
df_polls.head(10)
```

```python
df_polls.describe(include='all')
```

```python
# Finding out how many columns are there with no missing vals
na_count = dict(df_polls.isna().sum())
na_count
```

```python
# Making a list of these columns + state
clean_cols = [key for key, val in na_count.items() if val == 0]
clean_cols.append('state')
clean_cols
```

```python
# Creating a new dataframe with these columns
df_polls_clean = df_polls[clean_cols]
df_polls_clean.dropna(inplace = True)
df_polls_clean
```

```python
# 2020 cycle polls data
df_polls_hist = pd.read_csv("data/president_polls_historical.csv")
df_polls_hist.head(10)
```

```python
# Making a clean dataframe of the historical polls using the same columns
df_hist_clean = df_polls_hist[clean_cols]
df_hist_clean.dropna(inplace=True)
df_hist_clean
```

```python
# Merging the previous cycle and current cycle
df_merged = pd.concat([df_polls_clean,df_hist_clean], ignore_index = True)
df_merged
```

```python
# Pivoting the data to find the average voting percentage for all possible parties in each state
df_party = df_merged.pivot_table(index = ['state'], columns = 'party', values = 'pct').round(2)
df_party
```

```python
df_party.describe(include='all')
```

```python
# Since we only want significant parties, we clean up the data by removing those parties that have not been polled in at least 30 states
df_party.dropna(thresh = 30, inplace=True, axis=1)
df_party.fillna(0, inplace=True)
df_party
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler, normalize 
from sklearn.mixture import GaussianMixture 
from sklearn.metrics import silhouette_score 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn import metrics 
```

```python
# Storing the column headers
df_party_headers = df_party.columns

# Standardize data
scaler = StandardScaler() 
scaled_df = scaler.fit_transform(df_party) 

# Normalizing the Data 
normalized_df = normalize(scaled_df) 

# Converting the numpy array into a pandas DataFrame 
normalized_df = pd.DataFrame(normalized_df) 
normalized_df.columns = df_party_headers
normalized_df.rename_axis(None, axis = 1, inplace = True)
normalized_df
```

```python
# Applying the GMM clustering
gmm = GaussianMixture(n_components = 3) 
gmm.fit(df_party)
```

```python
# Visualizing the clustering 
for i in range(3):
    for j in range(i+1,4):
        plt.scatter(normalized_df[df_party_headers[i]], normalized_df[df_party_headers[j]],c = GaussianMixture(n_components = 3).fit_predict(normalized_df), \
            cmap = plt.cm.plasma, alpha = 0.5)
        plt.xlabel(df_party_headers[i])
        plt.ylabel(df_party_headers[j])
        plt.show() 
```

```python
# Getting the labels for the data using the predict function
labels = gmm.predict(df_party)
print(f"Cluster labels for all data:\n{labels}")
```

```python
# Adding labels to the dataframe
df_party['Labels'] = labels
df_party
```

```python
# Creating a new dataframe grouped by the labels, which gives our cluster data
df_gmm = df_party.groupby(by = 'Labels').mean().round(2)
df_gmm
```

**Notes**
- Based on the mean values, we can see that Label 0 is the cluster of Blue states, Label 1 is the cluster of Red states, and label 2 is the cluster of Swing states
- However, these labels change when you re-run the code, hence we have to adjust the values for creating the cluster dataframes each time
